**Question 5: What is the standard normal distribution, and why is it important?**

Answer 5: The standard normal distribution is a special type of normal distribution that has:

**Question 6: What is the Central Limit Theorem (CLT), and why is it critical in statistics?**

Answer 6: The Central Limit Theorem (CLT) states that when we take many random samples from any population, the sampling distribution of the sample mean will approximate a normal distribution, even if the original data is not normal ‚Äî provided the sample size is large enough (usually n ‚â• 30).

Importance:

It allows us to use normal distribution for making inferences about any population.

It helps in hypothesis testing and confidence interval estimation.

It makes statistical analysis reliable, even when data isn‚Äôt perfectly normal.

**Question 7: What is the significance of confidence intervals in statistical analysis?**

Answer 7: A confidence interval (CI) gives a range of values within which the true population parameter (like mean or proportion) is likely to fall, based on sample data.

Significance:

It shows how accurate or reliable a sample estimate is.

A wider interval means less certainty; a narrower interval means more precision.

Commonly used confidence levels are 95% or 99%, meaning we are that sure the true value lies in the range.

Example:
If the average height is 170 cm ¬± 5 cm at 95% confidence, we can say the true average height is between 165 and 175 cm.

**Question 9: Write a Python program to generate 1000 random numbers from a normal
distribution with mean = 50 and standard deviation = 5. Compute its mean and standard
deviation using NumPy, and draw a histogram to visualize the distribution.**

```python
#Answer 9:

import numpy as np
import matplotlib.pyplot as plt


data = np.random.normal(loc=50, scale=5, size=1000)


mean_value = np.mean(data)
std_value = np.std(data)

print("Mean:", mean_value)
print("Standard Deviation:", std_value)


plt.hist(data, bins=30, color='skyblue', edgecolor='black')
plt.title("Normal Distribution (Mean=50, SD=5)")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

**Question 10: You are working as a data analyst for a retail company. The company has
collected daily sales data for 2 years and wants you to identify the overall sales trend.
daily_sales = [220, 245, 210, 265, 230, 250, 260, 275, 240, 255,
235, 260, 245, 250, 225, 270, 265, 255, 250, 260]
‚óè Explain how you would apply the Central Limit Theorem to estimate the average sales
with a 95% confidence interval.
‚óè Write the Python code to compute the mean sales and its confidence interval.**

```python
#Answer 10: The Central Limit Theorem (CLT) says that if we take many random samples from a population, the sample means will follow a normal distribution, even if the data itself isn‚Äôt perfectly normal.So, we can use CLT to estimate the average daily sales and find a 95% confidence interval (CI) for the true mean.


import numpy as np
import scipy.stats as stats


daily_sales = [220, 245, 210, 265, 230, 250, 260, 275, 240, 255,
               235, 260, 245, 250, 225, 270, 265, 255, 250, 260]


mean_sales = np.mean(daily_sales)
std_error = stats.sem(daily_sales)


confidence = 0.95
ci = stats.t.interval(confidence, len(daily_sales)-1, loc=mean_sales, scale=std_error)

print("Mean Sales:", mean_sales)
print("95% Confidence Interval:", ci)
```
